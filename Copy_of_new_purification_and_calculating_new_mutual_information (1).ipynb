```python
import numpy as np
import itertools
```

```python
p = np.array([np.sqrt(1/2), np.sqrt(1/2)])              #plus   =A1
m = np.array([np.sqrt(1/2), -np.sqrt(1/2)])             #minus  =A2
pi = np.array([np.sqrt(1/2), 1j*np.sqrt(1/2)])          #plus_i =A3
mi = np.array([np.sqrt(1/2), -1j*np.sqrt(1/2)])        #minus_i =A4
```

```python
print(p,m,pi,mi, sep="\n")
```

```python
combination=[]
for i1 in [p,m,pi,mi]:
  for i2 in [p,m,pi,mi]:
    for i3 in [p,m,pi,mi]:
      combination.append([i1,i2,i3])
```

```python
len(combination)
combination
```

```python
#-------------   |X>   ---------------------------
X=[]
for i in combination:
  temp1=np.tensordot(i[0],i[1],axes=0).flatten()
  temp2=np.tensordot(temp1,i[2],axes=0).flatten()
  X.append(temp2[np.newaxis].T)


print(temp2[np.newaxis].T)       # 8 coloumns 1 row
print(temp2[np.newaxis].T.shape) # row matrix of size (8 X 1)
```
